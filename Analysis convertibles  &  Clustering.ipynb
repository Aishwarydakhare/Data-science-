```python
df['Mileage'] = df['Mileage'].str.replace(',', '')   ## To remove comma in Mileage column
```

# Clustering

K-Means Clustering

```python
feature_df = df[['Price', 'Mileage']]      ### Clustering on Price and Mileage
                
                
                
feature_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(feature_df)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(3)
kmeans.fit(feature_df)

identified_clusters = kmeans.fit_predict(feature_df)
identified_clusters

data_with_clusters = feature_df.copy()
data_with_clusters['Clusters'] = identified_clusters 
```

```python
fig = plt.figure(figsize=(15,10))
plt.scatter(data_with_clusters['Price'],data_with_clusters['Mileage'],c=data_with_clusters['Clusters'],cmap='rainbow')

centroids = kmeans.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]


plt.scatter(centroids_x,centroids_y,marker = "o", s=150,linewidths = 5, zorder = 10, c="black")
```

```python
kmeans = KMeans(3)
kmeans.fit(feature_df)

identified_clusters = kmeans.fit_predict(feature_df)
identified_clusters

data_with_clusters = feature_df.copy()
data_with_clusters['Clusters'] = identified_clusters 


fig = plt.figure(figsize=(15,10))
plt.scatter(data_with_clusters['Price'],data_with_clusters['MPG'],c=data_with_clusters['Clusters'],cmap='rainbow')

centroids = kmeans.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]


plt.scatter(centroids_x,centroids_y,marker = "o", s=150,linewidths = 5, zorder = 10, c="black")
```
